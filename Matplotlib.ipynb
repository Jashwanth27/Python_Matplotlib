```python
# Plot graph
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,7,5,3])
y=np.array([5,10,8,1])
plt.plot(x,y,color='g')
plt.show()
```

```python
# Plot Graph
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,11,1,1])
y=np.array([1,10,1,1])
plt.plot(x,y,ls=':',color='b')
plt.show()
```
